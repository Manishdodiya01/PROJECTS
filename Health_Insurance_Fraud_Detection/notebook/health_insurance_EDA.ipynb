```python
grouped_m = df.groupby(['policy_state', 'incident_type', 'fraud_reported']).agg({
    'months_as_customer': 'mean',
    'age': 'mean',
    'total_claim_amount': ['mean', 'sum'],
    # You can add more numerical columns here
})
```

```python
grouped_m.reset_index()
```

```python
import matplotlib.pyplot as plt

# Plot mean months_as_customer
grouped_m['months_as_customer']['mean'].plot(kind='bar', figsize=(10, 6))
plt.title('Mean Months as Customer by Group')
plt.xlabel('Group (Policy State, Incident Type, Fraud Reported)')
plt.ylabel('Mean Months as Customer')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot mean age
grouped_m['age']['mean'].plot(kind='bar', figsize=(10, 6))
plt.title('Mean Age by Group')
plt.xlabel('Group (Policy State, Incident Type, Fraud Reported)')
plt.ylabel('Mean Age')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Plot mean and sum total_claim_amount
grouped_m['total_claim_amount'].plot(kind='bar', figsize=(10, 6))
plt.title('Mean and Sum Total Claim Amount by Group')
plt.xlabel('Group (Policy State, Incident Type, Fraud Reported)')
plt.ylabel('Total Claim Amount')
plt.legend(['Mean', 'Sum'])
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
fraudulent_claims = df[df['fraud_reported'] == "Y"]
number_of_fraudulent_claims = len(fraudulent_claims)
total_claims = len(df['fraud_reported'])
fraud_rate = (number_of_fraudulent_claims / total_claims) * 100
print('FRAUD-RATE :',fraud_rate)
```

```python
number_of_fraudulent_claims
```

```python
policy_state_counts = df['policy_state'].value_counts()
total_claim_amount_fraudulent = fraudulent_claims['total_claim_amount']
age_fraudulent = fraudulent_claims['age']
incident_date = df['incident_date']  
incident_type = df['incident_type']
incident_severity = df['incident_severity']
insured_education_level = df['insured_education_level']
```

```python
# Pie Chart of Fraudulent vs. Non-Fraudulent Claims
plt.figure(figsize=(6, 6))
plt.pie([number_of_fraudulent_claims, total_claims - number_of_fraudulent_claims], labels=['Fraudulent', 'Non-Fraudulent'], autopct='%1.1f%%', colors=['red', 'blue'])
plt.title('Fraudulent vs. Non-Fraudulent Claims')
plt.show()
```

```python
# Bar Plot of Fraud Rate by Policy State
plt.figure(figsize=(8, 6))
sns.barplot(x=policy_state_counts.index, y=(policy_state_counts / total_claims * 100))
plt.xlabel('Policy State')
plt.ylabel('Fraud Rate (%)')
plt.title('Fraud Rate by Policy State')
plt.show()
```

```python
# Histogram of Total Claim Amount for Fraudulent Claims
plt.figure(figsize=(8, 6))
plt.hist(total_claim_amount_fraudulent, bins=20, color='red', alpha=0.7)
plt.xlabel('Total Claim Amount')
plt.ylabel('Frequency')
plt.title('Total Claim Amount Distribution for Fraudulent Claims')
plt.show()
```

```python
# Box Plot of Age for Fraudulent Claims
plt.figure(figsize=(8, 6))
sns.boxplot(y=age_fraudulent, color='red')
plt.ylabel('Age')
plt.title('Age Distribution for Fraudulent Claims')
plt.show()
```

```python
# Violin Plot of Total Claim Amount by Fraudulent vs. Non-Fraudulent Claims
plt.figure(figsize=(8, 6))
sns.violinplot(x=df['fraud_reported'], y=df['total_claim_amount'])
plt.xlabel('Fraudulent Claims')
plt.ylabel('Total Claim Amount')
plt.title('Total Claim Amount Distribution by Fraudulent vs. Non-Fraudulent Claims')
plt.show()
```

```python
# Scatter Plot of Age vs. Total Claim Amount for Fraudulent Claims
plt.figure(figsize=(8, 6))
plt.scatter(age_fraudulent, total_claim_amount_fraudulent, color='red')
plt.xlabel('Age')
plt.ylabel('Total Claim Amount')
plt.title('Age vs. Total Claim Amount for Fraudulent Claims')
plt.show()
```

```python
# Stacked Bar Plot of Fraudulent Claims by Education Level
plt.figure(figsize=(8, 6))
sns.countplot(x=insured_education_level, hue=df['fraud_reported'])
plt.xlabel('Education Level')
plt.ylabel('Count')
plt.title('Fraudulent Claims by Education Level')
plt.show()
```

```python
# Heatmap of Fraud Rate by Incident Type and Severity
incident_data = df.groupby(['incident_type', 'incident_severity']).size().unstack()
incident_data = incident_data.div(incident_data.sum(axis=1), axis=0) * 100  # Convert to percentages
plt.figure(figsize=(10, 6))
sns.heatmap(incident_data, annot=True, cmap='viridis')
plt.xlabel('Incident Severity')
plt.ylabel('Incident Type')
plt.title('Fraud Rate by Incident Type and Severity')
plt.show()
```
