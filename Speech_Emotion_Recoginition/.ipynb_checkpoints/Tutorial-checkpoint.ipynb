# Importing Libraries

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

#  Loading Audio Files

```python
audio_path = "discord-notification.wav"
audio_data , sr = librosa.load(audio_path)
```

# Extracting Features

```python
mfccs = librosa.feature.mfcc(y=audio_data, sr=sr)
```

# Visualizing Features

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time')
plt.colorbar()
plt.title('MFCC')
plt.tight_layout()
plt.show()
```

# Additional Preprocessing (Optional)

```python
mfccs_normalized = librosa.util.normalize(mfccs)
```
