```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.30,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

X_train_scaler = StandardScaler()
X_test_scaler = StandardScaler()
```

```python
X_train_scaled = X_train_scaler.fit_transform(X_train)
X_test_scaled = X_train_scaler.transform(X_test)
```

```python
X_test_scaled
```

```python
scaled_train_df = pd.DataFrame(X_train_scaled , columns=X_train.columns , index=X_train.index)
scaled_test_df = pd.DataFrame(X_test_scaled , columns=X_test.columns , index=X_test.index)
```

```python
scaled_train_df
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
```

```python
model_nb = GaussianNB()
model_rf = RandomForestClassifier()


model_nb.fit(scaled_train_df , y_train)
```

```python
model_rf.fit(scaled_train_df , y_train)
```

```python
from sklearn.metrics import confusion_matrix , classification_report , accuracy_score
```

```python
y_pred = model_nb.predict(scaled_test_df)

accuracy_score(y_test , y_pred)
```

```python
y_pred_rf = model_rf.predict(scaled_test_df)

accuracy_score(y_pred_rf , y_test)
```

```python
print(confusion_matrix(y_test , y_pred))
print(classification_report(y_test , y_pred))
```

```python
print(confusion_matrix(y_test , y_pred_rf))
print(classification_report(y_test , y_pred_rf))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {"var_smoothing" : [0.1,0.01,0.5,0.005,0.001,1e-6,1e-8,1e-10 , 1e-11]}
```

```python
grid = GridSearchCV(estimator=model_nb , param_grid=param_grid , cv=5 , verbose=3)
grid.fit(scaled_train_df , y_train)
```

```python
grid.best_params_
```

```python
best_grid = grid.best_estimator_
best_grid
```

```python
best_grid.fit(scaled_train_df , y_train)
```

```python
y_pred_new = best_grid.predict(scaled_test_df)

accuracy_score(y_pred_new , y_test)
```
